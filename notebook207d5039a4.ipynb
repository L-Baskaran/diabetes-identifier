```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('/kaggle/input/diabetes-dataset/diabetes.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.Outcome.value_counts()
```

```python
data.groupby('Outcome').mean()
```

```python
X=data.drop('Outcome',axis=1)
Y=data.Outcome
```

```python
scaler=StandardScaler()
```

```python
scaler_s=scaler.fit_transform(X)
```

```python
scaler_s
```

```python
X=scaler_s
Y=data.Outcome
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
x_train.shape
x_test.shape
```

```python
classa=svm.SVC(kernel='linear')
```

```python
classa.fit(x_train,y_train)
```

```python
x_a=classa.predict(x_train)
x_b=accuracy_score(x_a,y_train)
x_b
```

```python
x_c=classa.predict(x_test)
x_d=accuracy_score(x_c,y_test)
x_d
```
